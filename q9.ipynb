```python
#by using selenium and webdriver we are extract the data from website
import pandas as pd
import selenium
from selenium import webdriver
import time
```

```python
#we are using chrome browser as default
driver = webdriver.Chrome("Chromedriver.exe")
```

```python
#myntra website
url = 'https://www.myntra.com/'
driver.get(url)
```

```python
#we are searching for specific name
search_btn = driver.find_element_by_xpath("//div[@class='desktop-query']/input")
```

```python
#sending words to search
search_btn.send_keys("shoes")
```

```python
#sending words to search
search_btn = driver.find_element_by_xpath("//a[@class='desktop-submit']")
search_btn.click()
```

```python
url = 'https://www.myntra.com/shoes?f=Color:Black_36454f&plaEnabled=false&rf=Price:6612.0_13075.0_6612.0 TO 13075.0'
```

```python
driver.get(url)
```

```python
#urls
```

```python
urls = []
for i in range(0,9):
    for j in driver.find_elements_by_xpath('//li[@class="pagination-active"]/a'):
        urls.append(j.get_attribute("href"))
    driver.find_element_by_xpath('//li[@class="pagination-next"]/a').click()
    time.sleep(4)
                     
```

```python
#getting the url
urls
```

```python
shoe_name =[]
price_of_shoe = []
description_of_shoe = []
```

```python
for i in range(0,3):
    for j in driver.find_elements_by_xpath('//h3[@class="product-brand"]'):
        shoe_name.append(j.text)
    for k in driver.find_elements_by_xpath('//h4[@class="product-product"]'):
        description_of_shoe.append(k.text)
    for l in driver.find_elements_by_xpath('//div[@class="product-price"]'):
        price_of_shoe.append(l.text)
   
```

```python
#forming the dataframe
shoe=pd.DataFrame({})
shoe['shoe name']=shoe_name
shoe['price of shoe']=price_of_shoe
shoe['description of shoe']=description_of_shoe
```
