```python
#using webdriver i am going to extract the data from website
# i have used exception 
import selenium
import pandas as pd
import time

from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
#i have used google chrome
driver = webdriver.Chrome("Chromedriver.exe")
```

```python
#amazon website
driver.get('https://www.amazon.com/')
```

```python
#user put the product name to be searched
user_input = input('Enter the product to be searched')
```

```python
#the key word for searching

search_bar = driver.find_element_by_id('twotabsearchtextbox')
search_bar.clear()
search_bar.send_keys(user_input)
time.sleep(4)
search_button = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input')
search_button.click()      


```

```python
urls = []
for i in range(0,3):
    for j in driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]/a'):
        urls.append(j.get_attribute("href"))
```

```python
urls
```

```python
guitar_title=[]
guitar_price=[]
guitar_brand=[]
guitar_rate=[]
guitar_rate_total=[]
guitar_Availability=[]
guitar_description=[]
guitar_weight=[]
```

```python
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    
    try:
        jobs=driver.find_element_by_xpath("//h1[@class='a-size-large a-spacing-none']")
        guitar_title.append(jobs.text)
    except NoSuchElementException:
        guitar_title.append("-")
     
    try:
        #jobs=driver.find_element_by_xpath('//*[@id="priceblock_saleprice_row"]/td[2]')
        jobs=driver.find_element_by_xpath('//td[@class="a-span12"]')
        guitar_price.append(jobs.text)
    except NoSuchElementException:
        guitar_price.append("-")
       # //*[@id="priceblock_saleprice"]
    
    try:
        jobs=driver.find_element_by_xpath('//*[@id="poExpander"]/div[1]/div/table/tbody/tr[1]/td[2]/span')
        guitar_brand.append(jobs.text)
    except NoSuchElementException:
        guitar_brand.append("-")
        
    try:
        jobs=driver.find_element_by_xpath('//*[@id="acrCustomerReviewText"]')
        guitar_rate_total.append(jobs.text)
    except NoSuchElementException:
        guitar_rate_total.append("-")
    
    try:
        jobs=driver.find_element_by_xpath('//span[@class="a-size-medium a-color-base"]')
        guitar_rate.append(jobs.text)
    except NoSuchElementException:
        guitar_rate.append("-")
         
    try:
        jobs=driver.find_element_by_xpath('//div[@class="a-section a-spacing-base"]')
        guitar_Availability.append(jobs.text)
    except NoSuchElementException:
        guitar_Availability.append("-")  
        
    try:
        jobs=driver.find_element_by_xpath('//div[@class="disclaim"]')
        guitar_description.append(jobs.text)
    except NoSuchElementException:
        guitar_description.append("-") 
        
    
    try:
        jobs=driver.find_element_by_xpath('//td[@class="a-size-base prodDetAttrValue"]')
        guitar_weight.append(jobs.text)
    except NoSuchElementException:
        guitar_weight.append("-") 
    
    
    
```
