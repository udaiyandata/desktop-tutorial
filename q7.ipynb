```python
#by using selenium and webdriver we are extract the data from website
import pandas as pd
import selenium
from selenium import webdriver
```

```python
#we are using chrome browser as default
driver = webdriver.Chrome("Chromedriver.exe")
```

```python
#flipkart website
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace')
```

```python
#now we going to click the search button
search_button = driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']")
search_button.click()
```

```python
#getting all the url
urls = []
next_page = driver.find_elements_by_xpath('//nav[@class="yFHi8N"]/a')
for i in next_page:
    urls.append(i.get_attribute("href"))
```

```python
len(urls)
```

```python
rating=[]
review_summary=[]
full_review=[]
```

```python
#getting all the data base from the website 
for i in range(0,10):
    
        
    for j in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):
        rating.append(j.text)
    for k in driver.find_elements_by_xpath('//p[@class="_2-N8zT"]'):
        review_summary.append(k.text)
    for m in driver.find_elements_by_xpath('//div[@class="t-ZTKy"]'):
        full_review.append(m.text)
    
```
