```python
#using selenium for webscraping and also using the exception handling
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
##chromeweb browser is used
driver = webdriver.Chrome("Chromedriver.exe")
```

```python
#staticticstimes is used
url = 'http://statisticstimes.com/economy/india/indian-states-gdp.php'
driver.get(url)
```

```python
#state  is webscraped
state = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[2]')
state
```

```python
state1 = []
for i in state:
    state1.append(i.text)
state1
```

```python
rank = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[1]')
rank    
#rank  is webscraped
```

```python
rank1 = []
for i in rank:
    rank1.append(i.text)
rank1
```

```python
#gsdp1  is webscraped
GSDP1 = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[4]')
GSDP1 
```

```python
GSDP = []
for i in GSDP1:
    GSDP.append(i.text)
GSDP
```

```python
#share is webscraped
share = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[5]')
share
```

```python
share1 = []
for i in share:
    share1.append(i.text)
share1
```

```python
#gdp is webscraped
gdp= driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr/td[6]')
gdp
```

```python
gdp1 = []
for i in gdp:
    gdp1.append(i.text)
gdp1
```

```python
print(len(state1))
print(len(rank1))
print(len(GSDP))
print(len(share1))
print(len(gdp1))
```

```python
#dataframe is used
df = pd.DataFrame({})
df['rank'] = rank1
df['state'] = state1
df[' GSDP(18-19)'] = GSDP
df['share'] = share1
df['gdp1'] = gdp1 
```
