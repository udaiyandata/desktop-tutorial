```python
#using selenium for webscraping and also using the exception handling
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
#chromeweb browser is used
driver = webdriver.Chrome('Chromedriver.exe')
```

```python
#imdb website is used
url = 'https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
#name is webscraped
name = driver.find_elements_by_xpath('//*[@id="main"]/div/div[3]/div[3]/div/div[2]/h3/a')
```

```python
name
```

```python
name1 = []
for i in name:
    name1.append(i.text)
name1
```

```python
#year span is webscraped
year_span = driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')
```

```python
year_span1 = []
for i in year_span:
    year_span1.append(i.text)
year_span1
```

```python
#span is webscraped
span = driver.find_elements_by_xpath('//span[@class="genre"]')
```

```python
span1 = []
for i in span:
    span1.append(i.text)
span1
```

```python
#runtime is webscraped
runtime = driver.find_elements_by_xpath('//span[@class="runtime"]')
```

```python
runtime1 = []
for i in runtime:
    runtime1.append(i.text)
runtime1
```

```python
#rate is webscraped
rate = driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]')
```

```python
rate1 = []
for i in rate:
    rate1.append(i.text)
rate1
```

```python
#vote is webscraped
vote = driver.find_elements_by_name('nv')
```

```python
vote
```

```python
vote1 = []
for i in vote:
    vote1.append(i.text)
vote1
```

```python
print(len(name1))
print(len(year_span1))
print(len(span1))
print(len(runtime1))
print(len(rate1))
print(len(vote1))
```

```python
#data frame is used
df = pd.DataFrame({})
df['name'] = name1
df['year_span'] = year_span1
df['span'] = span1
df['runtime'] = runtime1
df['rate1'] = rate1
df['vote'] = vote1
```
