```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating")
#page2=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
first_title=soup.find_all('h3',class_="lister-item-header")
first_title
```

```python
top_movie=[]
for i in first_title:
    top_movie.append(i.text.replace('\n',''))
top_movie
```

```python
second_title=soup.find_all('span',class_="lister-item-year text-muted unbold")
second_title
```

```python
top_movie_year=[]
for i in second_title:
    top_movie_year.append(i.text.replace('\n',''))
top_movie_year
```

```python
third_title=soup.find_all('div' , class_='inline-block ratings-imdb-rating')
third_title
```

```python
top_rate=[]
for i in third_title:
     top_rate.append(i.text.replace('\n',''))
top_rate
    
```

```python
print(len(top_movie),len(top_movie_year),len(top_rate))
```

```python
import pandas as pd
j=pd.DataFrame({})
j['top_movie']=top_movie
j['top_movie_year']=top_movie_year
j['top_rate']=top_rate
j.to_csv("data1.csv")
```
