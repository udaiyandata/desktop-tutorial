```python
second_head=soup.find_all('h2',class_='mp-h2')
second_head
```

```python
second_name= []

for i in second_head:
    second_name.append(i.get_text().replace('\n',''))
second_name
```

```python
len(second_name)
```

```python
third_head=soup.find_all('h3',class_='vector-menu-heading')
third_head
```

```python
third_name=[]
for i in third_head:
    third_name.append(i.get_text().replace('\n',''))
third_name
```
