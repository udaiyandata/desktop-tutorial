```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/search/?utf8=%E2%9C%93&query=gener")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
author_title=soup.find_all('h4' ,class_='italic')
author_title
```

```python
a=[]

for i in author_title:
    a.append(i.get_text().replace('\n',''))
    
a
```

```python
author_genre=soup.find_all('p' ,class_='genre-links hidden-phone')
author_genre
```

```python
b=[]

for i in author_genre:
    b.append(i.get_text().replace('\n',''))
    
b
```

```python
author_words=soup.find_all('p' ,class_='sans')
author_words
```

```python
c=[]

for i in author_words:
    c.append(i.get_text().replace('\n',''))
    
c
```
